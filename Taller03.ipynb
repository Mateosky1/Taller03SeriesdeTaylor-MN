## Ejercicio 1: $f(x) = \cos(x)$ alrededor de $x_0 = 0$

### 1. Desarrollo Matemático
Evaluamos la función y sus derivadas en el punto central $x_0 = 0$:

* $f(x) = \cos(x) \implies f(0) = 1$
* $f'(x) = -\sin(x) \implies f'(0) = 0$
* $f''(x) = -\cos(x) \implies f''(0) = -1$
* $f'''(x) = \sin(x) \implies f'''(0) = 0$
* $f^{(4)}(x) = \cos(x) \implies f^{(4)}(0) = 1$

Reemplazando en la fórmula de Maclaurin, obtenemos los polinomios (los órdenes impares se anulan):
* **Orden 0 y 1:** $P_1(x) = 1$
* **Orden 2 y 3:** $P_3(x) = 1 - \frac{x^2}{2}$
* **Orden 4:** $P_4(x) = 1 - \frac{x^2}{2} + \frac{x^4}{24}$

### 2. Estimaciones (Evaluación del método)
Vamos a estimar el valor del coseno en dos puntos cercanos a cero para ver la precisión:

* **Estimación 1 (en $x = 0.5$):**
  * Valor Real: $\cos(0.5) \approx 0.8775$
  * Con Orden 2: $P_2(0.5) = 1 - \frac{0.5^2}{2} = 0.8750$
* **Estimación 2 (en $x = 1.0$):**
  * Valor Real: $\cos(1.0) \approx 0.5403$
  * Con Orden 2: $P_2(1.0) = 1 - \frac{1^2}{2} = 0.5000$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 400)
y_real = np.cos(x)
y_p0_p1 = np.ones_like(x)
y_p2_p3 = 1 - (x**2) / 2
y_p4 = 1 - (x**2) / 2 + (x**4) / 24

plt.figure(figsize=(10, 5))
plt.plot(x, y_real, 'k', linewidth=2, label='Real: cos(x)')
plt.plot(x, y_p0_p1, '--', label='Orden 0 y 1')
plt.plot(x, y_p2_p3, '--', label='Orden 2 y 3')
plt.plot(x, y_p4, '--', label='Orden 4')
plt.scatter([0], [1], color='red', zorder=5, label='Centro x0=0')
plt.title('Aproximaciones de Taylor: cos(x)')
plt.ylim(-2, 2)
plt.xlim(-4, 4)
plt.grid(True, linestyle=':')
plt.legend()
plt.show()

# Imprimir estimaciones
print("--- ESTIMACIONES PARA COS(X) ---")
print(f"Estimación en x=0.5 -> Real: {np.cos(0.5):.4f} | Orden 2: {1 - (0.5**2)/2:.4f} | Orden 4: {1 - (0.5**2)/2 + (0.5**4)/24:.4f}")
print(f"Estimación en x=1.0 -> Real: {np.cos(1.0):.4f} | Orden 2: {1 - (1.0**2)/2:.4f} | Orden 4: {1 - (1.0**2)/2 + (1.0**4)/24:.4f}")
```

## Ejercicio 2: $f(x) = \frac{1}{1-x}$ alrededor de $x_0 = 0$

### 1. Desarrollo Matemático
Reescribimos la función como $f(x) = (1-x)^{-1}$ y derivamos usando la regla de la cadena:

* $f(x) = (1-x)^{-1} \implies f(0) = 1$
* $f'(x) = (1-x)^{-2} \implies f'(0) = 1$
* $f''(x) = 2(1-x)^{-3} \implies f''(0) = 2$
* $f'''(x) = 6(1-x)^{-4} \implies f'''(0) = 6$
* $f^{(4)}(x) = 24(1-x)^{-5} \implies f^{(4)}(0) = 24$

Sustituyendo y simplificando con los factoriales correspondientes:
* **Orden 0:** $P_0(x) = 1$
* **Orden 1:** $P_1(x) = 1 + x$
* **Orden 2:** $P_2(x) = 1 + x + x^2$
* **Orden 4:** $P_4(x) = 1 + x + x^2 + x^3 + x^4$

### 2. Estimaciones (Evaluación del método)
* **Estimación 1 (en $x = 0.2$):**
  * Valor Real: $\frac{1}{1-0.2} = 1.2500$
  * Con Orden 2: $P_2(0.2) = 1 + 0.2 + 0.04 = 1.2400$
* **Estimación 2 (en $x = 0.5$):**
  * Valor Real: $\frac{1}{1-0.5} = 2.0000$
  * Con Orden 4: $P_4(0.5) = 1 + 0.5 + 0.25 + 0.125 + 0.0625 = 1.9375$

```python
x2 = np.linspace(-0.9, 0.9, 400)
y2_real = 1 / (1 - x2)
y2_p0 = np.ones_like(x2)
y2_p1 = 1 + x2
y2_p2 = 1 + x2 + x2**2
y2_p4 = 1 + x2 + x2**2 + x2**3 + x2**4

plt.figure(figsize=(10, 5))
plt.plot(x2, y2_real, 'k', linewidth=2, label='Real: 1/(1-x)')
plt.plot(x2, y2_p0, '--', label='Orden 0')
plt.plot(x2, y2_p2, '--', label='Orden 2')
plt.plot(x2, y2_p4, '--', label='Orden 4')
plt.scatter([0], [1], color='red', zorder=5, label='Centro x0=0')
plt.title('Aproximaciones de Taylor: 1/(1-x)')
plt.ylim(0, 6)
plt.xlim(-1, 1)
plt.grid(True, linestyle=':')
plt.legend()
plt.show()

# Imprimir estimaciones
print("--- ESTIMACIONES PARA 1/(1-X) ---")
print(f"Estimación en x=0.2 -> Real: {1/(1-0.2):.4f} | Orden 2: {1 + 0.2 + 0.2**2:.4f} | Orden 4: {1 + 0.2 + 0.2**2 + 0.2**3 + 0.2**4:.4f}")
print(f"Estimación en x=0.5 -> Real: {1/(1-0.5):.4f} | Orden 2: {1 + 0.5 + 0.5**2:.4f} | Orden 4: {1 + 0.5 + 0.5**2 + 0.5**3 + 0.5**4:.4f}")
```

## Ejercicio 3: $f(x) = \ln(x)$ alrededor de $x_0 = 1$

### 1. Desarrollo Matemático
En este caso, la función logaritmo no puede evaluarse en cero, por lo que expandimos la serie alrededor de $x_0 = 1$:

* $f(x) = \ln(x) \implies f(1) = 0$
* $f'(x) = x^{-1} \implies f'(1) = 1$
* $f''(x) = -x^{-2} \implies f''(1) = -1$
* $f'''(x) = 2x^{-3} \implies f'''(1) = 2$
* $f^{(4)}(x) = -6x^{-4} \implies f^{(4)}(1) = -6$

Armamos los polinomios reemplazando el término $x$ por $(x-1)$ según dicta la fórmula:
* **Orden 0:** $P_0(x) = 0$
* **Orden 1:** $P_1(x) = (x-1)$
* **Orden 2:** $P_2(x) = (x-1) - \frac{(x-1)^2}{2}$
* **Orden 4:** $P_4(x) = (x-1) - \frac{(x-1)^2}{2} + \frac{(x-1)^3}{3} - \frac{(x-1)^4}{4}$

### 2. Estimaciones (Evaluación del método)
* **Estimación 1 (en $x = 1.2$):**
  * Valor Real: $\ln(1.2) \approx 0.1823$
  * Con Orden 2: $P_2(1.2) = (0.2) - \frac{0.2^2}{2} = 0.1800$
* **Estimación 2 (en $x = 1.5$):**
  * Valor Real: $\ln(1.5) \approx 0.4054$
  * Con Orden 2: $P_2(1.5) = (0.5) - \frac{0.5^2}{2} = 0.3750$

```python
x3 = np.linspace(0.1, 2.5, 400)
y3_real = np.log(x3)
y3_p0 = np.zeros_like(x3)
y3_p1 = x3 - 1
y3_p2 = (x3 - 1) - 0.5*(x3 - 1)**2
y3_p4 = (x3 - 1) - 0.5*(x3 - 1)**2 + (1/3)*(x3 - 1)**3 - 0.25*(x3 - 1)**4

plt.figure(figsize=(10, 5))
plt.plot(x3, y3_real, 'k', linewidth=2, label='Real: ln(x)')
plt.plot(x3, y3_p0, '--', label='Orden 0')
plt.plot(x3, y3_p2, '--', label='Orden 2')
plt.plot(x3, y3_p4, '--', label='Orden 4')
plt.scatter([1], [0], color='red', zorder=5, label='Centro x0=1')
plt.title('Aproximaciones de Taylor: ln(x)')
plt.ylim(-2, 1.5)
plt.xlim(0, 2.5)
plt.grid(True, linestyle=':')
plt.legend()
plt.show()

# Imprimir estimaciones
print("--- ESTIMACIONES PARA LN(X) ---")
print(f"Estimación en x=1.2 -> Real: {np.log(1.2):.4f} | Orden 2: {(1.2-1) - ((1.2-1)**2)/2:.4f} | Orden 4: {(1.2-1) - ((1.2-1)**2)/2 + ((1.2-1)**3)/3 - ((1.2-1)**4)/4:.4f}")
print(f"Estimación en x=1.5 -> Real: {np.log(1.5):.4f} | Orden 2: {(1.5-1) - ((1.5-1)**2)/2:.4f} | Orden 4: {(1.5-1) - ((1.5-1)**2)/2 + ((1.5-1)**3)/3 - ((1.5-1)**4)/4:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Algoritmo del Teorema de Lagrange (basado en sumatorias y productorias)
def polinomio_lagrange(x_datos, y_datos, x_evaluar):
    n = len(x_datos)
    resultado = np.zeros_like(x_evaluar)
    for k in range(n):
        L_k = np.ones_like(x_evaluar)
        for i in range(n):
            if i != k:
                L_k = L_k * (x_evaluar - x_datos[i]) / (x_datos[k] - x_datos[i])
        resultado = resultado + (y_datos[k] * L_k)
    return resultado
```

## PARTE 2: Polinomios de Lagrange

### Ejercicio 1: Datos $(0,0), (30,0.5), (60, \frac{\sqrt{3}}{2}), (90,1)$

**1. Identificación de puntos:**
Tenemos $n=4$ puntos. Como el primer punto es $(0,0)$, el valor de $y_0 = 0$, lo que anula todo el primer término de la sumatoria. Solo necesitamos calcular $L_1(x)$, $L_2(x)$ y $L_3(x)$.

**2. Cálculo de los polinomios base $L_k(x)$:**
$$L_1(x) = \frac{(x-0)(x-60)(x-90)}{(30-0)(30-60)(30-90)} = \frac{x(x-60)(x-90)}{54000}$$

$$L_2(x) = \frac{(x-0)(x-30)(x-90)}{(60-0)(60-30)(60-90)} = \frac{x(x-30)(x-90)}{-54000}$$

$$L_3(x) = \frac{(x-0)(x-30)(x-60)}{(90-0)(90-30)(90-60)} = \frac{x(x-30)(x-60)}{162000}$$

**3. Polinomio final $P(x)$:**
Multiplicamos cada base por su respectivo $y_k$:
$$P(x) = 0.5 \left[ \frac{x(x-60)(x-90)}{54000} \right] + \frac{\sqrt{3}}{2} \left[ \frac{x(x-30)(x-90)}{-54000} \right] + 1 \left[ \frac{x(x-30)(x-60)}{162000} \right]$$

```python
x1 = np.array([0, 30, 60, 90])
y1 = np.array([0, 0.5, np.sqrt(3)/2, 1])

x1_cont = np.linspace(-10, 100, 200)
y1_cont = polinomio_lagrange(x1, y1, x1_cont)

plt.figure(figsize=(8, 5))
plt.plot(x1_cont, y1_cont, 'b-', linewidth=2, label='Polinomio de Lagrange')
plt.scatter(x1, y1, color='red', s=80, zorder=5, label='Puntos dados')
plt.title('Lagrange: Datos Trigonométricos')
plt.grid(True, linestyle=':')
plt.legend()
plt.show()
```

### Ejercicio 2: Datos $(1,1), (2,2), (3,2)$

**1. Identificación de puntos:**
Tenemos $x_0=1, y_0=1$; $x_1=2, y_1=2$; $x_2=3, y_2=2$.

**2. Cálculo de los polinomios base $L_k(x)$:**
$$L_0(x) = \frac{(x-2)(x-3)}{(1-2)(1-3)} = \frac{x^2 - 5x + 6}{2}$$

$$L_1(x) = \frac{(x-1)(x-3)}{(2-1)(2-3)} = \frac{x^2 - 4x + 3}{-1} = -x^2 + 4x - 3$$

$$L_2(x) = \frac{(x-1)(x-2)}{(3-1)(3-2)} = \frac{x^2 - 3x + 2}{2}$$

**3. Construcción y simplificación de $P(x)$:**
$$P(x) = 1 \cdot L_0(x) + 2 \cdot L_1(x) + 2 \cdot L_2(x)$$
$$P(x) = \frac{x^2 - 5x + 6}{2} + 2(-x^2 + 4x - 3) + 2\left(\frac{x^2 - 3x + 2}{2}\right)$$
$$P(x) = 0.5x^2 - 2.5x + 3 - 2x^2 + 8x - 6 + x^2 - 3x + 2$$

Agrupando términos semejantes, obtenemos la parábola final:
$$P(x) = -0.5x^2 + 2.5x - 1$$
